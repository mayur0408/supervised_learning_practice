```python
import numpy as np 
import pandas as pd 
df=pd.read_csv("Car details v3.csv")
df.head()
```

```python
df.info()
```

```python
df=df.dropna()
```

```python
df.info()
```

```python
len(df['name'].unique())
```

```python
carname_encode=df.groupby(df['name'])['selling_price'].mean().sort_values(ascending=True).to_dict()
```

```python
carname_encode
```

```python
df['name']=df['name'].map(carname_encode)
df.head()
```

```python
df['fuel'].unique()
```

```python
dummy1=pd.get_dummies(df['fuel'],drop_first=True)
dummy1.head()
```

```python
df=pd.concat([df,dummy1],axis=1)
df.head()
```

```python
df['seller_type'].unique()
```

```python
dummy2=pd.get_dummies(df['seller_type'],drop_first=True)
dummy2.head()
```

```python
df=pd.concat([df,dummy2],axis=1)
df.head()
```

```python
df["transmission"].unique()
```

```python
dummy3=pd.get_dummies(df['transmission'],drop_first=True)
dummy3.head()
```

```python
df=pd.concat([df,dummy3],axis=1)
df.head()
```

```python
df['owner'].unique()
```

```python
owner_encode={'First Owner':4,'Second Owner':3,'Third Owner':2,'Fourth & Above Owner':1,'Test Drive Car':0}
print(owner_encode)
```

```python
df['owner']=df['owner'].map(owner_encode)
df.head(10)
```

```python
df['mileage']=df['mileage'].str.replace("[a-z/]","")
```

```python
df.head(10)
```

```python
df['mileage']=pd.to_numeric(df['mileage'])
```

```python
df.info()
```

```python
df['engine']=df['engine'].str.replace("CC","")
df['engine']=pd.to_numeric(df['engine'])
df.head()
```

```python
df.info()
```

```python
df['max_power']=df['max_power'].str.replace("bhp","")
df['max_power']=pd.to_numeric(df['max_power'])
df.head()
```

```python
df.info()
```

```python
df=df.drop(['name','fuel','seller_type','transmission','torque'],axis=1)
df.head()
```

```python
import seaborn as sns 
sns.heatmap(df.corr())
```

```python
sns.lmplot(x="engine",y="selling_price",data=df)
```

```python
sns.lmplot(x="year",y="selling_price",data=df,row="owner",hue="owner")
```

```python
X=df.drop(['selling_price'],axis=1)
X.head()
```

```python
Y=df['selling_price']
Y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,random_state=42)
print(f"size of traning data is {len(X_train)} {len(Y_train)}")
print(f"size of test data is {len(X_test)} {len(Y_test)}")
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()
model.fit(X_train,Y_train)
```

```python
Y_pred=model.predict(X_test)
from sklearn.metrics import r2_score
score=r2_score(Y_test,Y_pred)
print(score)
```

```python
model.score(X_test,Y_test)
```

```python
from sklearn.model_selection import cross_val_score
print(cross_val_score(model,X,Y,cv=10,scoring='r2').mean())
```
